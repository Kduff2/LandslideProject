#### Figure 1: Landslides per month

```python
import pandas as pd
import matplotlib.pyplot as plt

LS = pd.read_csv("nasa_US_landslides.csv")

Months = LS['Months']
Landslides = LS['Landslides per month']
Month_Names = LS['Month_Names']

plt.bar(Months, Landslides)
plt.xticks(Months, labels=Month_Names)
plt.xlabel("Months")
plt.ylabel("Landslide Count")
plt.title("Landslide Count per Month")
```

#### Figure 2: Landslide Fatality count per month

```python
Months = LS['Months']
Fatalities = LS['Death per month']
Month_Names = LS['Month_Names']

plt.bar(Months, Fatalities)
plt.xticks(Months, labels=Month_Names)
plt.xlabel("Months")
plt.ylabel("Fatality Count")
plt.title("Fatality Count per Month")
```

#### Figure 3: Landslide count per year

```python
Year = LS['Years']
Landslides = LS['Landslides per year']

plt.bar(Year, Landslides)
plt.xlabel("Years")
plt.ylabel("Landslide Count")
plt.title("Landslide Count per Year")
```

#### Figure 5: Map of all events from shapefile

```python

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import geopandas as gpd

landslide_events = gpd.read_file("nasa_coolr_reports_point.shp")

landslide_events.head()

global_landslides = plt.subplot(1, 1, 1, projection=ccrs.PlateCarree())

global_landslides.add_feature(cfeature.COASTLINE)
global_landslides.add_feature(cfeature.BORDERS)

plt.plot(landslide_events.longitude, landslide_events.latitude, '.', transform=ccrs.PlateCarree())
```

#### Firgure 6: Map of landslide events in the US

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import geopandas as gpd


landslides = plt.subplot(1, 1, 1, projection=ccrs.AlbersEqualArea(central_longitude=-99, central_latitude=35))

landslides.set_extent([-125, -65, 20, 50])

landslides.add_feature(cfeature.COASTLINE)
landslides.add_feature(cfeature.BORDERS)
landslides.add_feature(cfeature.STATES)

landslide_events = gpd.read_file("nasa_coolr_reports_point.shp")
plt.plot(landslide_events.longitude, landslide_events.latitude, '.', transform=ccrs.PlateCarree())
plt.title('US Landslides')
plt.legend

```

```python
import xarray as xr

data = xr.open_dataset('precip.V1.0.mon.ltm.1991-2020.nc')

data
```

```python
precip = data['precip']
max_precip = precip.max()
max_precip
```

```python

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

max_precip = precip.max(dim='time')

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.AlbersEqualArea(central_longitude=-99, central_latitude=35))

# Add features to the map
ax.add_feature(cfeature.LAND, edgecolor='black')
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.LAKES, edgecolor='black')

# Add the precipitation data
precipitation_plot = max_precip.plot(ax=ax, transform=ccrs.PlateCarree(), cmap='rainbow', 
                                            cbar_kwargs={'label': 'Precipitation (mm)'})

# Set title and labels
ax.set_title('Maximum Precipitation')

plt.show()
```
