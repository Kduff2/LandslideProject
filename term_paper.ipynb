#### Figure 1: Landslides per month

#### Mapping Precipitation Data

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import xarray as xr

data = xr.open_dataset('precip.V1.0.mon.ltm.1991-2020.nc')
lat = data.variables['lat'][:]
lon = data.variables['lon'][:]
time = data.variables['time'][:]
precip = data.variables['precip'][:]


map_precip = plt.subplot(1, 1, 1, projection=ccrs.AlbersEqualArea(central_longitude=-99, central_latitude=35))

map_precip.set_extent([-125, -65, 20, 50])


map_precip.add_feature(cfeature.OCEAN)
map_precip.add_feature(cfeature.COASTLINE)
map_precip.add_feature(cfeature.STATES, linestyle=':')
map_precip.add_feature(cfeature.BORDERS, linestyle=':')
map_precip.add_feature(cfeature.LAKES, edgecolor='Black')



plt.contourf(lon, lat, precip[0, :, :], transform=ccrs.PlateCarree(), cmap='rainbow')
plt.colorbar(label='Precipitation (mm)')
plt.title('Precipitation')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

#### Maximum values from Precipitation Data

```python
import numpy as np
precip = data.variables['precip'][:]

max_precip = np.max(precip)

print(max_precip)
```

```python
import numpy as np
max_precip = np.max(precip, axis=0)

fig = plt.figure(figsize=(10, 6))
ax = plt.axes(projection=ccrs.AlbersEqualArea(central_longitude=-99, central_latitude=35))

contour = ax.contourf(lon, lat, max_precip, cmap='rainbow', transform=ccrs.PlateCarree())
plt.colorbar(contour, ax=ax, label='Precipitation (mm)')

ax.add_feature(cfeature.LAND, edgecolor='black')
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAKES, edgecolor='black')

ax.set_title('Maximum Precipitation')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```

```python
import numpy as np
min_precip = np.min(precip, axis=0)

fig = plt.figure(figsize=(10, 6))
ax = plt.axes(projection=ccrs.AlbersEqualArea(central_longitude=-99, central_latitude=35))

contour = ax.contourf(lon, lat, min_precip, cmap='rainbow', transform=ccrs.PlateCarree())
plt.colorbar(contour, ax=ax, label='Precipitation (mm)')

ax.add_feature(cfeature.LAND, edgecolor='black')
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAKES, edgecolor='black')

ax.set_title('Minimum Precipitation')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```
